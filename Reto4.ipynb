1. Dado un número entero, determinar si ese número corresponde al código ASCII de una vocal minúscula.

A partir de la tabla del código ASCII vemos que los códigos de las letras minúsculas son lo números que estan entre 97,122.

```python
n = int(input("Ingrese un número entero"))
if n >= 97 and n <= 122:
    print("El número ingresado corresponde al código ASCII de una vocal minúscula")
else:
    print("El número ingresado no corresponde al código ASCII de una vocal minúscula")
```

2. Dada una cadena de longitud 1, determine si el código ASCII de primera letra de la cadena es par o no.

```python
c = str(input("Ingrese un carácter: "))
if len(c) != 1:
  print("Error: La cadena debe tener exactamente una letra.")
if ord(c) % 2 == 0:
    print("El código ASCII de la primera letra es par")
else:
    print("El código ASCII de la primera letra es impar")
```

3. Dado un carácter, construya un programa en Python para determinar si el carácter es un dígito o no.

```python
c = str(input("Ingrese un carácter: "))
if c.isdigit():
    print(f"'{c}' es un dígito.")
else:
    print(f"'{c}' no es un dígito.")
```

4. Realice un programa que lea dos números reales y determine si el primero es múltiplo del segundo.

```python
a = float(input("Ingrese el primer número real: "))
b = float(input("Ingrese el segundo número real: "))

if a % b == 0:
    print(f"{a} es múltiplo de {b}")
else:
    print(f"{a} no es múltiplo de {b}")
```

6. Dado el centro y el radio de un círculo, determinar si un punto de R2 pertenece o no al interior del círculo.

```python
center_x = float(input("Ingrese el valor de x del centro del círculo: "))
center_y = float(input("Ingrese el valor de y del centro del círculo: "))
r = float(input("Ingrese el radio del círculo: "))

x = float(input("Ingrese el valor de x del punto: "))
y = float(input("Ingrese el valor de y del punto: "))

d = (x - center_x)**2 + (y - center_y)**2

if d <= r**2:
    print("El punto pertenece al interior del círculo.")
else:
    print("El punto no pertenece al interior del círculo.")
```

7. Dadas tres longitudes positivas, determinar si con esas longitudes se puede construir un triángulo.

```python
a = float(input("Ingrese la primera longitud: "))
b = float(input("Ingrese la segunda longitud: "))
c = float(input("Ingrese la tercera longitud: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    print("Se puede construir un triángulo con las longitudes dadas.")
else:
    print("No se puede construir un triángulo con las longitudes dadas.")
```
